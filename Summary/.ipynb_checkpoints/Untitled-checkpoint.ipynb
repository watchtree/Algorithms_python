```python
def bubbleSort(alist):
    length = len(alist)
    for j in range(length - 1):
        for i in range(0, length - j -1):
            if alist[i] > alist[i + 1]:
                alist[i] , alist[i + 1] = alist[i + 1], alist[i]
    return
```

```python
li = [54,26,93,17,77,31,44,55,20]
bubbleSort(li)
print(li)
```

```python
#
def bubbleSort(alist):
    length = len(alist)
    for j in range(length - 1):
        count = 0
        for i in range(0, length - j -1):
            if alist[i] > alist[i + 1]:
                alist[i] , alist[i + 1] = alist[i + 1], alist[i]
                count += 1
        if count == 0:
            break
    return
```

```python
li = [54,26,93,17,77,31,44,55,20]
bubbleSort(li)
print(li)
```

Quicksortpartition-exchange sort



1. ""pivot
2. partition
3. recursive

iteration



```python
def quickSort(alist, start, end):
    if len(alist) <= 1 or start == end:
        return
    oriStart = start
    oriEnd = end
    temp = alist[start]
    while start < end:
        while start < end and alist[end] >= temp:
            end -= 1
        alist[start] = alist[end]
        while start < end and alist[start] < temp:
            start += 1
        alist[end] = alist[start]
    alist[start] = temp
    quickSort(alist, oriStart, start)
    quickSort(alist, start+1, oriEnd)
```

```python
li = [1,3,2,26,93,17,17,77,31,44,55,20]
quickSort(li,0 , len(li)-1)
print(li)
```

## 

Selection sort

nn-1




* O(n2) 
* O(n2) 
*  

```python
def selectSort(alist):
    if alist == []:
        return 
    length = len(alist)
    for i in range(length - 1):
        min = i
        for j in range(i, length):
            if alist[j] < alist[min]:
                min = j
        alist[i], alist[min] = alist[min], alist[i]
        
```

```python
li = [1,3,2,26,93,17,17,77,31,44,55,20]
selectSort(li)
print(li)
```

```python
def heap_sort(L):
    L_length = len(L) - 1
    first_sort_count = L_length//2
    
    #(heap_adjust)
    for i in range(first_sort_count):
        head_adjust(L, first_sort_count - i, L_length)
    
    #(swap_param)(heap_adjust)
    for i in range(L_length -  1):
        L = swap_param(L, 1, L_length - i)
        head_adjust(L, 1, L_length - i -1)
    
    return [L[i] for i in range(1, len(L))]
```

```python
#(swap_param)
def swap_param(L, i, j):
    L[i], L[j] = L[j], L[i]
    return L
```

```python
#
def head_adjust(L, start, end):
    temp = L[start]
    i = start
    j = 2 * i
    
    while j <= end:
        #
        if j < end and L[j]<L[j+1]:
            #j
            j += 1
        if temp < L[j]:
            #
            L[i] = L[j]
            i = j
            j = 2 * i
        else:
            break
    L[i] = temp
    
```

```python
L = deque([50, 16, 30, 10, 60,  90,  2, 80, 70])
L.appendleft(0)
heap_sort(L)
```
